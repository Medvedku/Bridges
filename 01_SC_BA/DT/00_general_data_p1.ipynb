```python
import requests


def get_elevation(lat, lng):
    """
    Function to get elevation in meters from OpenTopoData API.
    
    :param lat: Latitude of the point.
    :param lng: Longitude of the point.
    :return: Elevation in meters or None if not available.
    """
    opentopodata_base_url = "https://api.opentopodata.org/v1/eudem25m"
    params = {
        "locations": f"{lat},{lng}"
    }

    try:
        response = requests.get(opentopodata_base_url, params=params)
        response.raise_for_status()  # Raise an error for bad responses
        data = response.json()

        # Check if the response contains an elevation result
        if "results" in data and len(data["results"]) > 0:
            # Extract the elevation from the first result
            elevation = data["results"][0]["elevation"]
            return elevation
        else:
            print("No elevation data found for the specified location.")
            return None
    except requests.RequestException as e:
        print(f"Request failed: {e}")
        return None


# # Example usage
# latitude = 48.790152  # E.g., Latitude for the Eiffel Tower
# longitude = 17.123319  # E.g., Longitude for the Eiffel Tower
# elevation = get_elevation(latitude, longitude)
# print(f"Elevation at {latitude}, {longitude}: {elevation} meters")
```

```python
import time

# Function to update elevations for NaN values


def update_elevations(df):
    # Open log file for writing
    log_file = open("elevation_update_log.txt", "w")
    for index, row in df.iterrows():
        if pd.isna(row["Výška"]):  # Check if elevation is NaN
            # Call the get_elevation function
            elevation = get_elevation(
                row["Zemepisná_šírka"], row["Zemepisná_dĺžka"])
            # Update elevation if available
            if elevation is not None:
                df.at[index, "Výška"] = elevation
                # Print and write to log file
                message = f"Elevation updated for index {index} (ID_mosta: {row['ID_mosta']}): {
                    elevation} meters\n"
                print(message)
                log_file.write(message)
            else:
                # Print and write to log file
                message = f"No elevation data found for index {
                    index} (ID_mosta: {row['ID_mosta']}).\n"
                print(message)
                log_file.write(message)
            # Delay for 1 second between requests
            time.sleep(1)
    log_file.close()  # Close the log file after writing


# Call the function to update elevations
update_elevations(df_bridges_all)
```

```python
def check_for_nezisteny(df, column_name):
    """
    Check if there are any occurrences of "Nezistený" in the specified column of the DataFrame.

    Args:
        df (DataFrame): The DataFrame to search.
        column_name (str): The name of the column to check.

    Returns:
        int: The count of occurrences of "Nezistený" in the specified column.
    """
    # Filter the DataFrame to find occurrences of "Nezistený" in the specified column
    nezisteny_occurrences = df[df[column_name] == "Nezistený"]

    # Return the count of occurrences
    return nezisteny_occurrences.shape[0]


def check_for_nezisteny_in_all_columns(df):
    """
    Iterate through all columns of the DataFrame and check for occurrences of "Nezistený".

    Args:
        df (DataFrame): The DataFrame to search.

    Returns:
        dict: A dictionary containing the column names as keys and the count of occurrences of "Nezistený" as values.
    """
    nezisteny_counts = {}
    for column_name in df.columns:
        count_nezisteny = check_for_nezisteny(df, column_name)
        nezisteny_counts[column_name] = count_nezisteny
    return nezisteny_counts


# Example usage:
nezisteny_counts = check_for_nezisteny_in_all_columns(df_bridges_all)
for column_name, count_nezisteny in nezisteny_counts.items():
    if count_nezisteny > 0:
        print(f"There are {count_nezisteny} occurrences of 'Nezistený' in column '{
              column_name}'.")
    else:
        pass
```

```python
import pandas as pd
import plotly.express as px

# Assuming df_M023 is your DataFrame
average_set = "12H"

# Convert 'time_server_UTC' to datetime and resample
df_M023['time_server_UTC'] = pd.to_datetime(df_M023['time_server_UTC'])

# Resample and compute mean for the relevant columns
df_temp_1 = df_M023.resample(average_set, on='time_server_UTC')[
    '21.pv1'].mean()
df_temp_2 = df_M023.resample(average_set, on='time_server_UTC')[
    '21.pv2'].mean()

# Calculate the difference between temperatures
temperature_difference = df_temp_1 - df_temp_2

# Interpolate missing values
temperature_difference_interpolated = temperature_difference.interpolate()

# Sort temperature_difference_interpolated by index
temperature_difference_interpolated = temperature_difference_interpolated.sort_index()

# Calculate min and max differences
min_difference = temperature_difference_interpolated.min()
max_difference = temperature_difference_interpolated.max()

# Create a Plotly line plot for temperature differences with interpolation
fig_temperature_difference = px.line()
fig_temperature_difference.add_scatter(
    x=temperature_difference_interpolated.index, y=temperature_difference_interpolated, name='Teplotný rozdiel', mode='lines', line_shape='spline', line_color='#333333')

fig_temperature_difference.update_layout(title='Teplotný spád',
                                         xaxis_title='Dátum',
                                         yaxis_title='Teplotný rozdiel [°C]',
                                         legend=dict(orientation='h', yanchor='bottom',
                                                     y=1.02, xanchor='right', x=1),
                                         plot_bgcolor='#EEEEEE',
                                         width=900, height=500,
                                         margin=dict(l=50, r=25, t=50, b=50))

# Add text annotations for min and max differences
fig_temperature_difference.add_annotation(
    text=f'Minimálny rozdiel: {min_difference:.1f} °C', xref='paper', yref='paper', x=0.5, y=0.95, showarrow=False)
fig_temperature_difference.add_annotation(
    text=f'Maximálny rozdiel: {max_difference:.1f} °C', xref='paper', yref='paper', x=0.5, y=0.90, showarrow=False)

# Show the plot
fig_temperature_difference.show()
```

```python
import pandas as pd

# Replace the file path with your actual file path
file_path = r'C:\Users\relia\Documents\GitHub\Bridges\Data\sr_co_mostydc_reportpk_2024-01-01.csv'

# Import CSV file as DataFrame with specified encoding
df_2024 = pd.read_csv(file_path, encoding='windows-1250')

# Now df_2024 contains your data from the CSV file
```

```python
# Define a function to remove the suffix from the ID
def remove_suffix(id):
    return id.split('.')[0] if isinstance(id, str) else id


# Apply the function to the "ID DC" column
df_2024['ID DC'] = df_2024['ID DC'].apply(remove_suffix)
```

```python
df_2024.columns
```

```python
# Define a function to check if a given ID is valid based on your criteria
def is_valid_id(id_value):
    return isinstance(id_value, str) and id_value.startswith('M') and id_value[1:].isdigit()

# Function to find a valid ID from four columns and replace if needed


def replace_invalid_id(row):
    # Check if 'ID mosta' is valid
    if not is_valid_id(row['ID mosta']):
        # Attempt to replace from 'ID DC' if valid
        if is_valid_id(row['ID DC']):
            return row['ID DC']
        # Next, try 'správcovské číslo' if valid
        elif is_valid_id(row['správcovské číslo']):
            return row['správcovské číslo']
        # Then check 'Stavebno-technický stav' if valid
        elif is_valid_id(row['Stavebno-technický stav']):
            return row['Stavebno-technický stav']
        else:
            # Return original if no valid replacement found
            return row['ID mosta']
    else:
        return row['ID mosta']  # Return original if it is already valid


# Filter for entries where 'Most /DC' equals "Most"
df_most = df_2024[df_2024['Most /DC'] == "Most"].copy()

# Apply the function to replace invalid 'ID mosta'
df_most['ID mosta'] = df_most.apply(replace_invalid_id, axis=1)

# Update the main DataFrame with cleaned 'ID mosta' values
df_2024.update(df_most['ID mosta'])
```

```python
# Set the values in the "ID DC" column to NaN for rows where df_2024['Most /DC'] == "Most"
df_2024.loc[df_2024['Most /DC'] == "Most", 'ID DC'] = None
```

```python
df_2024
```

```python
# Split the DataFrame into two separate DataFrames based on the values in the "Most /DC" column
df_bridges = df_2024[df_2024['Most /DC'] == 'Most'].copy()[["ID mosta"]]
df_DC = df_2024[df_2024['Most /DC'] == 'Dilatačný celok'].copy()

# Remove duplicates from the "ID mosta" column
df_bridges = df_bridges.drop_duplicates(subset=["ID mosta"])
```

```python
df_bridges["ID mosta"].value_counts(), df_DC["ID DC"].value_counts()
```

```python
# Define the mapping of conditions to their respective values
condition_mapping = {
    'Zlý': 5,
    'Veľmi zlý': 6,
    'Uspokojivý': 4,
    'Bezchybný': 1,
    'Dobrý': 3,
    'Havarijný': 7,
    'Nezistený': 0,
    'Veľmi dobrý': 2
}

# Map the values from 'Stavebno-technický stav' column to 'n_2024' based on the condition_mapping
df_DC['n_2024'] = df_DC['Stavebno-technický stav'].map(condition_mapping)
```

```python
# Iterate over each row in df_bridges
for index, row in df_bridges.iterrows():
    id_mosta = row["ID mosta"]
    # Find matching rows in df_DC
    matching_rows = df_DC[df_DC["ID DC"] == id_mosta]

    # Calculate the mean of "n_2024" values for the matching rows
    mean_n_2024 = matching_rows["n_2024"].mean()

    # Assign the mean value to the corresponding row in df_bridges
    df_bridges.at[index, "n_2024"] = mean_n_2024
```

```python
bridges_path = r"C:\Users\relia\Documents\GitHub\Bridges\Data\df_bridges_data.csv"
df_bridges_all = pd.read_csv(bridges_path, encoding='utf-8', delimiter=',')
df_bridges_all.columns
```

```python
# Create a dictionary from df_bridges mapping 'ID mosta' to 'n_2024'
id_to_condition_mapping = df_bridges.set_index('ID mosta')['n_2024'].to_dict()

# Map this dictionary to df_bridges_all['ID_mosta'] to create a new 'n_2024' column
df_bridges_all['n_2024'] = df_bridges_all['ID_mosta'].map(
    id_to_condition_mapping)

df_bridges_all
```

```python
# First, filter df_bridges to only include rows with a known condition
known_condition_bridges = df_bridges.dropna(subset=['n_2024'])

# Convert the ID columns to sets for easier comparison
set_known_bridges = set(known_condition_bridges['ID mosta'])
set_bridges_all = set(df_bridges_all['ID_mosta'].dropna())

# Find bridges in known_condition_bridges that are not in df_bridges_all
missing_known_bridges = set_known_bridges - set_bridges_all

# Count these bridges
print(f'Number of bridges with known condition in df_bridges not found in df_bridges_all: {
      len(missing_known_bridges)}')
```

```python
# Convert the set to a list for easier use in filtering
missing_known_bridges_list = list(missing_known_bridges)

# Create a mask that checks if 'ID mosta' or 'ID DC' contains any of the IDs in missing_known_bridges_list
mask = df_2024['ID mosta'].isin(
    missing_known_bridges_list) | df_2024['ID DC'].isin(missing_known_bridges_list)

# Filter df_2024 using the mask to create a new DataFrame
filtered_df_2024 = df_2024[mask]

# Display the filtered DataFrame to check the results
print(filtered_df_2024[['ID mosta', 'ID DC']].head())
```

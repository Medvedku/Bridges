```python
import pandas as pd

# Specify the path to your CSV file
csv_file_path = 'Data/surveys/Survey.csv'

# Load the CSV file into a DataFrame
df_survey = pd.read_csv(csv_file_path)

df_survey
```

```python
# Create a copy of the DataFrame
df_copy = df_survey.copy()

# Convert 'CESTA' column to numeric and coerce non-convertible values to NaN
df_copy['CESTA'] = pd.to_numeric(df_copy['CESTA'], errors='coerce')

# Drop rows with NaN values in 'CESTA' column
df_copy = df_copy.dropna(subset=['CESTA'])

# Sort the copy by 'S' in descending order
df_copy_sorted = df_copy.sort_values(by='S', ascending=False)

# Get the first row for each unique OKRES in the sorted copy
df_first_by_okres = df_copy_sorted.groupby('OKRES').head(3)

# Reset the 'CESTA' column back to integers in the resulting DataFrame (if needed)
df_first_by_okres['CESTA'] = df_first_by_okres['CESTA'].astype(int)
```

```python
# Set the maximum number of rows to display
pd.set_option('display.max_rows', None)

# Print the DataFrame
print(df_first_by_okres)

# Reset the maximum number of rows to the default value (optional)
pd.reset_option('display.max_rows')
```

```python
import pandas as pd

# Define the file path
# nodes_path = "Data/Nodes_processed.csv"
# roads_path = "Data/Roads_processed.csv"
bridges_path = "Data/Bridges_processed.csv"

# Load the CSV data into a DataFrame
# df_nodes = pd.read_csv(nodes_path, encoding='utf-8', delimiter=',')
# df_roads = pd.read_csv(roads_path, encoding='utf-8', delimiter=',')
df_bridges = pd.read_csv(bridges_path, encoding='utf-8', delimiter=',')
```

```python
df_bridges.columns, df_bridges["Okres"].value_counts().keys()
```

```python
import folium
import pandas as pd

# Assuming you have a DataFrame named df_bridges containing bridge data
okresino = "Bratislava I"

# Filter bridges in the "Prešov" okres and meet the condition n_2023 >= 5
presov_bridges = df_bridges[(df_bridges["Okres"] == okresino) & (
    df_bridges["n_2023"] >= 5)]

# Create a map centered around the Prešov okres
map_center = [49.0000, 21.2500]  # Adjust the center as needed
mymap = folium.Map(location=map_center, zoom_start=6)

tile = folium.TileLayer(
    tiles='https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
    attr='Esri',
    name='Esri Satellite',
    overlay=False,
    control=True
).add_to(mymap)

# Iterate through the filtered bridges and add markers to the map
for index, bridge in presov_bridges.iterrows():
    bridge_location = [bridge["Zemepisná_šírka"], bridge["Zemepisná_dĺžka"]]
    bridge_id = bridge["ID_mosta"]
    bridge_class = bridge["Trieda_PK"]
    bridge_name = bridge["Názov_mosta"]
    bridge_n_2023 = bridge["n_2023"]

    tooltip_text = f"ID Mosta: {bridge_id}<br>Trieda PK: {bridge_class}<br>Názov Mosta: {bridge_name}<br>n_2023: {bridge_n_2023}"
    popup_text = f"ID Mosta: {bridge_id}<br>Trieda PK: {bridge_class}<br>Názov Mosta: {bridge_name}<br>n_2023: {bridge_n_2023}"

    folium.Marker(
        bridge_location,
        tooltip=tooltip_text,
        # Clickable pop-up with max width
        popup=folium.Popup(popup_text, max_width=300),
    ).add_to(mymap)


# Display the map
mymap.save('presov_bridges_map.html')  # Save the map to an HTML file
```

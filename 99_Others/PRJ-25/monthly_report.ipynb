```python
from datetime import datetime, timezone

# Define the dates
date_from = datetime(2025, 2, 15, 0, 0, 0, tzinfo=timezone.utc)
date_to = datetime(2025, 3, 15, 0, 0, 0, tzinfo=timezone.utc)

# Convert to epoch timestamps
epoch_from = int(date_from.timestamp())
epoch_to = int(date_to.timestamp())

# Print the results
print("Epoch from:", epoch_from)
print("Epoch to:", epoch_to)
```

```python
# Create the MongoDB aggregation pipeline
pipeline = [
    {
        "$match": {
            "time.server.epoch": {
                "$gte": epoch_from,
                "$lte": epoch_to
            }
        }
    },
    {
        "$sort": {
            "time.server.epoch": 1
        }
    }
]
```

```python
# Fetch documents
documents = list(collection.aggregate(pipeline))

# Initialize DataFrames
hub_data = {
    "124005320765128": [],  # HUB for Piers 03 & 04
    "229128940302024": [],  # HUB for Piers 05 & 06
    "277769444929224": []   # HUB for Abutment
}
```

```python
# Iterate through documents and structure data
for doc in documents:
    uuid = str(doc["meta"]["uuid"])  # Convert UUID to string for lookup
    timestamp = doc["time"]["server"]["epoch"]
    voltage = doc["meta"]["power"]["battery"]["voltage"]

    # Find corresponding pier mapping
    pier = next((p for p, info in pier_info.items()
                if info["UUID"] == uuid), None)
    if pier is None:
        continue  # Skip if UUID is not in pier_info

    sensor_data = {"Time": timestamp, "Voltage": voltage}

    # Extract sensor values
    for sensor_id, sensor_info in pier_info[pier]["Sensors"].items():
        sensor_type = sensor_info["Type"]
        sensor_values = doc["values"].get(str(sensor_id), {})

        if sensor_type == "TLeaf":
            sensor_data[f"T_{sensor_id}_pv1"] = sensor_values.get("pv1", None)
            sensor_data[f"T_{sensor_id}_pv2"] = sensor_values.get("pv2", None)
        elif sensor_type == "PLeaf":
            sensor_data[f"P_{sensor_id}_pv0"] = sensor_values.get("pv0", None)
            sensor_data[f"P_{sensor_id}_pv1"] = sensor_values.get("pv1", None)
            sensor_data[f"P_{sensor_id}_pv3"] = sensor_values.get("pv3", None)
        elif sensor_type == "ILeaf":
            sensor_data[f"I_{sensor_id}_pv0"] = sensor_values.get("pv0", None)
            sensor_data[f"I_{sensor_id}_pv1"] = sensor_values.get("pv1", None)
            sensor_data[f"I_{sensor_id}_pv3"] = sensor_values.get("pv3", None)

    # Store data in corresponding HUB
    hub_data[uuid].append(sensor_data)

# Convert lists to DataFrames
df_hub_1 = pd.DataFrame(hub_data["124005320765128"])  # HUB for Piers 03 & 04
df_hub_2 = pd.DataFrame(hub_data["229128940302024"])  # HUB for Piers 05 & 06
df_hub_3 = pd.DataFrame(hub_data["277769444929224"])  # HUB for Abutment
```
